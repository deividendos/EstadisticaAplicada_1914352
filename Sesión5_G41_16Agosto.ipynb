```python
#Primer paso pandas
#Llammar las librerías a utilizar

import pandas as pd
import numpy as np
```

```python
calificaciones = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/Calificaciones.csv')
```

```python
calificaciones
```

```python
calificaciones.info
```

```python
#Renombrar
calificaciones = calificaciones.rename(columns={'Calficación Final' : 'CalFinal'})
```

```python
calificaciones = calificaciones.loc[:, calificaciones.columns!='Id']
```

```python
calificaciones.head()
```

```python
#Promedio del grupo
calificaciones['CalFinal'].mean()
```

```python
#Contar alumnos
calificaciones['Nombre'].count()
```

```python
#Etiquetar Aprobados y reprobados
calificaciones['Situación'] = np.where(calificaciones['CalFinal']>=70, 'Aprobado', 'Reprobado')
```

```python
calificaciones
```

```python
#promedio de aprobados y promedio de reporobados
calificaciones.groupby('Situación').mean()
```

```python
#Contar aprobados y reprobados 
#usar función aggregate agg
calificaciones.groupby('Situación').agg('count')
```

```python
#OTRA FORMA
#Contar aprobados y reprobados 
#usar función aggregate agg
calificaciones.groupby('Situación')['CalFinal'].agg('count')
```

```python
#OTRA FORMA
#Contar aprobados y reprobados 
#usar función aggregate agg
calificaciones.groupby('Situación')['CalFinal'].agg(['count', 'mean', 'std']).round(1)
```

## Librería Matplotlib

Es el graficador estándar de phyton 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
personas = [1, 2, 3, 4]
edades = [28, 21, 30, 27]
```

```python
##graficar personas por edad 'plt.plot(lista de variables para eje x, lista de variables para eje y)
plt.plot(personas,edades)
plt.show()
```

```python
## Para grafica de dispersión
plt.scatter(personas,edades, color = ['blue', 'red', 'yellow', 'green']) # color es para poner colores a cada eemento de la lista
plt.xlabel('personas') # Para ponerle nombre a el eje x
plt.ylabel('edad (años)') #para ponerle nombre al eje y
plt.title('Edades de 4 desconocidos') # para ponerle titulo al grafico
plt.show() # para quitar el url
```

```python
#Grafica de barras
plt.bar(personas, edades)
```
