```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
nacimientos = pd.DataFrame({'Municipios' : ['Mty', 'Apodaca', 'Escobedo', 'Santiago'], 'bb_nacidos' : [233, 569, 780, 123], 'bb_moridos' : [123, 80, 23, 12]})
```

```python
nacimientos
```

```python
plt.bar(nacimientos['Municipios'], nacimientos['bb_nacidos'])
```

```python
nacimientos = nacimientos.sort_values('bb_nacidos', ascending=False)
#Ordenar el dataframe de mayor a menor para que la grafica salga ordenada de la misma manera
```

```python
plt.bar(nacimientos['Municipios'], nacimientos['bb_nacidos'])
plt.title('Nacimientos en N.L. en 2020')
plt.xlabel('municipios')
plt.ylabel('cantidad de bebés')
plt.show()
```

```python
Personas = ['Juan', 'Paola', 'Ana', 'Alejandra']
edades = [23, 18, 20, 31]
```

```python
plt.bar(Personas, edades)
```

```python
#fig es el nombre del lienzo
#(panel1, panel2) son los argumentos de nuestro lienzo fig, donde cada uno es una grafica, pueden ser mas de 2 graficas
# =plt.subplots(2,1) es para acomodar nuestras n graficas, en este caso las va a acomodar en 2 fiolas y 1 columna.
#Es decir, es plt.subplots(fila, columna)
fig, (panel1, panel2) = plt.subplots(2,1)
fig.tight_layout()

#grafica1 nacimientos
x1 = nacimientos['Municipios']
y1 = nacimientos['bb_nacidos']

panel1.bar(x1, y1)

#grafica2 edades
x2 = Personas
y2 = edades

panel2.bar(x2, y2)
```
